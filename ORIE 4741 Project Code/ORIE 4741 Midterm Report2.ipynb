```julia
x1 = Array{Float64}(undef, 7, 2)
for i = 1:7
    x1[i] = i^2
end

x2 = Array{Float64}(undef, 7, 2)
for i = 1:7
    x2[i] = i^3
end

X1 = [copy(x1[:,1]) ones(length(j))]
X2 = [copy(x2[:,1]) ones(length(j))]

w1 = X1\y
w2 = X2\y

h1 = w1[1]*x1
h2 = w2[1]*x2

c = h + h1 + h2
p = h + h1
```

```julia
function MSE(r, pred)
    mean(abs.(pred - r))
end

MSE_linear = MSE(y, h[:, 1])
MSE_2 = MSE(y, p[:, 1])
MSE_cubic = MSE(y, c[:, 1])


println("Linear MSE\t", MSE_linear)
println("Linear MSE\t", MSE_2)
println("Cubic MSE \t", MSE_cubic)
```
